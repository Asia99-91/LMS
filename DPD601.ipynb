```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("store.csv" , encoding="latin-1")
```

```python
# Assuming the DataFrame is already loaded and named df
# Ensure correct data types
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Average Order Value (AOV)
aov = df['Sales'].sum() / df['Order ID'].nunique()

# Sales by Category
sales_by_category = df.groupby('Category')['Sales'].sum()

# Profit by Region
profit_by_region = df.groupby('Region')['Profit'].sum()

# Top Products by Sales
top_products_by_sales = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

# Top Customers by Revenue
top_customers_by_revenue = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)

# Discount Impact Analysis
discount_impact = df.groupby('Discount')['Profit'].sum()

# Profit Margin
df['Profit Margin'] = df['Profit'] / df['Sales']
average_profit_margin = df['Profit Margin'].mean()


# Sales by Category
sales_by_category.plot(kind='bar', figsize=(10, 6), title="Sales by Category")
plt.ylabel('Sales ($)')
plt.show()

# Profit by Region
profit_by_region.plot(kind='bar', figsize=(10, 6), title="Profit by Region", color='orange')
plt.ylabel('Profit ($)')
plt.show()

# Top Products by Sales
top_products_by_sales.plot(kind='bar', figsize=(10, 6), title="Top Products by Sales", color='green')
plt.ylabel('Sales ($)')
plt.show()

# Top Customers by Revenue
top_customers_by_revenue.plot(kind='bar', figsize=(10, 6), title="Top Customers by Revenue", color='purple')
plt.ylabel('Revenue ($)')
plt.show()

# Discount Impact Analysis
discount_impact.plot(kind='bar', figsize=(10, 6), title="Discount Impact on Profit", color='red')
plt.ylabel('Profit ($)')
plt.show()

# Print the results
print(f"Average Order Value (AOV): {aov}")
print("\nSales by Category:\n", sales_by_category)
print("\nProfit by Region:\n", profit_by_region)
print("\nTop Products by Sales:\n", top_products_by_sales)
print("\nTop Customers by Revenue:\n", top_customers_by_revenue)
print("\nDiscount Impact Analysis:\n", discount_impact)


```
