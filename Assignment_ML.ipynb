```python
avg_price_zipcode=df.groupby('zipcode')['price'].mean().sort_values(ascending=False)
avg_price_zipcode.head()
```

```python
avg_price_bedrooms = df.groupby('bedrooms')['price'].mean().sort_values(ascending=False)
avg_price_bedrooms
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title('Number of Bedrooms vs. Property Prices')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='waterfront', y='price', data=df)
plt.title('Waterfront Property vs. Prices')
plt.xlabel('Waterfront (0 = No, 1 = Yes)')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['sqft_living'], df['price'], alpha=0.5)

plt.title('Relationship between Living Area Size (sqft_living) and Price')
plt.xlabel('Living Area Size (sqft_living)')
plt.ylabel('Price')

plt.show()
```

```python
avg_price_condition = df.groupby('condition')['price'].mean()

plt.figure(figsize=(8, 5))
avg_price_condition.plot(kind='bar', color='c', alpha=0.7)

plt.title('Average Property Price by Condition')
plt.xlabel('Condition (1 = Poor, 5 = Excellent)')
plt.ylabel('Average Price')
plt.show()
```

```python
avg_price_per_year = df.groupby('yr_built')['price'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(avg_price_per_year['yr_built'], avg_price_per_year['price'], color='g')

plt.title('Average Property Price Over Years')
plt.xlabel('Year Built')
plt.ylabel('Average Price')



plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['sqft_lot15'], df['price'], alpha=0.5, color='orange')

plt.title('Relationship between Lot Size in Nearby Area (sqft_lot15) and Price')
plt.xlabel('Lot Size in Nearby Area (sqft_lot15)')
plt.ylabel('Price')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

hb = plt.hexbin(df['long'], df['lat'], gridsize=50, C=df['price'], reduce_C_function=np.mean, cmap='coolwarm', mincnt=1)

plt.title('Property Prices by Geographic Location (Hexbin)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add a color bar to show price scale
cb = plt.colorbar(hb)
cb.set_label('Average Price')

plt.show()
```

```python
avg_price_floors = df.groupby('floors')['price'].mean()

plt.figure(figsize=(8, 5))
avg_price_floors.plot(kind='bar', color='blue', alpha=0.7)

plt.title('Average Property Price by Number of Floors')
plt.xlabel('Number of Floors')
plt.ylabel('Average Price')
plt.show()
```
